### Exercise 1. 

Read Total Profit for all months and show it using a line plot with Month Number as your baseline

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.style.use('seaborn-whitegrid')
sales = pd.read_csv('./data/company_sales_data.csv', sep = ',')
sales.head(11)
````

````python
plt.plot(sales['month_number'],sales['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.show()
````

### Exercise 2. 

Get Total Units Sold of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Units Sold Number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
````

````python
plt.plot(sales['month_number'], sales['total_units'], linewidth = 3, linestyle = 'dotted', marker = 'o', color = 'red')
plt.xlabel('Month Number')
plt.ylabel('Units Sole Number')
plt.title('Total Profit of All Months')
plt.legend(['Total Profit Over Time'], loc = 'lower right')
plt.xticks(np.arange(0,12.1, step = 2))
plt.yticks(np.arange(15000,45001, step = 5000))
plt.xlim(0,12)
plt.show()
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
````

````python
plt.plot(sales['month_number'], sales['facecream'], linestyle = 'dotted', c = 'black')
plt.plot(sales['month_number'], sales['facewash'], c = 'blue')
plt.plot(sales['month_number'], sales['toothpaste'], c = 'red')
plt.plot(sales['month_number'], sales['bathingsoap'], linestyle = '-.', c = 'green')
plt.plot(sales['month_number'], sales['shampoo'], c = 'purple', marker = 'o')
plt.plot(sales['month_number'], sales['moisturizer'], c = 'purple', linestyle = '--', marker = '^', mec = 'orange', markerfacecolor = 'orange')
plt.title('All Product Sales')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.xlim(0,12)
plt.legend(['Facecream','Facewash','Toothpaste','Bathing Soap','Shampoo','Moisturizer'], loc = 'upper left')
plt.show()
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
sales.plot.bar(x = 'month_number', y = ['facewash','facecream'], color = ['blue','yellow'])
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.xlim(0,12)
plt.ylim(0,3999)
plt.legend(['Facewash','Facecream'], loc = 'upper left', prop = {'size':9})
plt.show()
````
