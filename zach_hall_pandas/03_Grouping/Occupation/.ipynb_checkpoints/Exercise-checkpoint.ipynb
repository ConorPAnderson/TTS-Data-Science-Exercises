### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean().sort_values()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def m_f_ratio(item):
    if item == 'M':
        return 1
    else:
        return 0

users['male_ratio'] = users['gender'].apply(m_f_ratio) 

male_percent = users.groupby('occupation').male_ratio.sum() / users.occupation.value_counts() * 100

male_percent.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(['occupation']).age.agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.agg(['min','max'])
```

### Step 8.  For each occupation present the percentage of women and men

```python
def m_f_ratio(item):
    if item == 'M':
        return 1
    else:
        return 0

users['male_ratio'] = users['gender'].apply(m_f_ratio)
users['male_percent'] = users.male_ratio.sum() / users.occupation.value_counts() * 100
users['female_percent'] = 100 - users.male_percent
users[['male_percent', 'female_percent']]
```
